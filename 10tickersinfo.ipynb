```python
import pandas as pd
```

```python
data = pd.read_csv('/content/nasdaq100_stock_data.csv', header=[0,1])
```

**Reasoning**:
The subtask is to install the yfinance library. I will use pip to install it.



```python
tickers = data.columns.get_level_values(0)
filtered_tickers = tickers[1:-1]
unique_tickers = filtered_tickers.unique()
import yfinance as yf

ticker_info_list = []

for ticker_symbol in unique_tickers:
    try:
        ticker = yf.Ticker(ticker_symbol)
        info = ticker.info
        ticker_info = {
            'ticker': info.get('symbol', ticker_symbol),
            'name': info.get('longName', 'N/A')
        }
        ticker_info_list.append(ticker_info)
    except Exception as e:
        print(f"Could not fetch info for ticker {ticker_symbol}: {e}")
ticker_info_df = pd.DataFrame(ticker_info_list)
ticker_info_df
```
